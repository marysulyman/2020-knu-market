```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_excel('rnd_contest_data.xlsx', sheet_name="REF raw data")
br_data = data[data["Brand"] == "LG"]
models = br_data["Model"]
source1 = data[data["Source"]=="Traditional Sales"]
source2 = data[data["Source"]=="Internet Sales"]
```

```python
for i in models:
    br_data = source2[source2["Model"]==i]
    time = br_data["Month"]
    sales = br_data["Sales Units"]
    plt.ylabel = ('price')
    plt.xlabel = ('sales units')
    plt.plot(time,sales)
    plt.suptitle('Sales Dynamics')
    
plt.show()
```
